```python
import sqlite3

connection = sqlite3.connect('books.db')
```

```python
import pandas as pd

pd.options.display.max_columns = 10
pd.read_sql('select last from authors order by last desc', connection)
```

```python
pd.read_sql('select title from titles order by title desc', connection)
```

```python
pd.read_sql('''
    select titles.title, titles.copyright, titles.isbn
    from titles
    inner join author_ISBN on titles.isbn = author_ISBN.isbn
    inner join authors on author_ISBN.id = authors.id
    where authors.last = 'Deitel'
    order BY titles.title ASC
''', connection)
```

```python
cursor = connection.cursor()
cursor = cursor.execute(''' insert into authors (first, last)
values ('John', 'Doe')''')
```
